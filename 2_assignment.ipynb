### 16- Write a function called analyze_car_data that has at least two parameters as follows:
- Count can be price or count or power
- Property can be City, Brand, Fuel, Transmission, Chassis orColor.

**If you call the function as follows then it will return the average car price of each city.

**If you call the function as follows then it will return the number of cars for each transmission type.

**If you call the function as follows then it will return the average power of cars for each Brand.

```python
def analyze_car_data(count_type, property_type):
    if count_type!="Count":
        return dict(df.groupby(property_type)[count_type].mean())
    else:
        return dict(df.groupby(property_type)[property_type].count())
```
